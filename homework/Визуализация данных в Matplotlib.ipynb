```С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.```

````python
t = np.linspace(0, 10, 51)
t
````

````python
f = np.sin(t)
f
````

````python
fig, ax = plt.subplots()
ax.plot(t, f, color='green')
ax.set_xlim([0.5, 9.5])
ax.set_ylim([-2.5, 2.5])

title_font = {'fontsize':20, 'color':'DarkBlue'}
axis_font = {'fontsize':16, 'color':'black'}
plt.title('График f(t)', fontdict=title_font)
plt.ylabel('Значения f', fontdict=axis_font)
plt.xlabel('Значения x', fontdict=axis_font)
plt.show()
````

# Задание 3
```С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.```

````python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
````

````python
fig, ax = plt.subplots(2, 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('Графки y1 = x ** 2')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2 = 2 * x + 0.5')

ax3.plot(x, y3)
ax3.set_title('График y3 = -3 * x - 1.5')

ax4.plot(x, y4)
ax4.set_title('График y4 = np.sin(x)')

fig.set_size_inches(8, 6)
fig.subplots_adjust(hspace=0.3, wspace=0.3)

plt.show()
````

Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “Class”.

````python
import pandas as pd
plt.style.use('fivethirtyeight')
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
````

````python
dataset = pd.read_csv('creditcard.csv', sep=',')
````

````python
dataset.describe()
````

````python
dataset.head()
````

````python
dataset['Class'].value_counts().plot(kind='bar')
````

````python
dataset['Class'].value_counts().plot(kind='bar', logy=True)
````

````python
info_v1_fraud = dataset.loc[dataset['Class'] == 1, 'V1']
info_v1_normal = dataset.loc[dataset['Class'] == 0, 'V1']

fig, ax = plt.subplots(1, 2)
ax1, ax2 = ax.flatten()

ax1.hist(info_v1_fraud, bins=20, ec='black', density=True)
ax1.set_title('Fraud')

ax2.hist(info_v1_normal, bins=20, ec='black', density=True)
ax2.set_title('Normal')

fig.set_size_inches(6, 3)
fig.subplots_adjust(hspace=0.3, wspace=0.4)
````

````python
hists = ['Class 1', 'Class 0']
colors = ['red', 'grey']
n_bins = 20
alpha = 0.5
plt.hist(info_v1_fraud, label='Class 1', 
         bins=n_bins, color='red', density=True, alpha=0.6)
plt.hist(info_v1_normal, label='Class 0', 
         bins=n_bins, color='grey', density=True, alpha=0.6)
plt.legend()
plt.xlabel('Class')
````

````python
target_mean = round(dataset['V24'].mean(), 2)
target_median = dataset['V24'].median()
target_mode = dataset['V24'].mode()[0]
````
