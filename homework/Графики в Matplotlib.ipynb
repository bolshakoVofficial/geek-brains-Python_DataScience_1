```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 
           'Jupiter', 'Saturn', 'Uranus', 'Neptune']

masses = [0.055274, 0.815, 1.0, 0.107, 
          317.8, 95.0, 14.6, 17.147]
```

```python
plt.plot(planets, masses)
plt.show()
```

```python
plt.plot(planets, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info.plot()
plt.yscale('log')
plt.show()
```

```python
planet_info.plot(kind='bar')
plt.show()
```

```python
planet_info.plot(kind='bar', logy=True)
plt.show()
```

```python
planet_info.plot(kind='barh', logx=True)
plt.show()
```

```python
url = "http://infographics.economist.com/2018/databank/BMFile2000toJan2018.xls"
bmi = pd.read_excel(url)
```

```python
bmi.shape
```

```python
bmi.iloc[:, :6].head(10)
```

```python
bmi_price = bmi['dollar_price']
bmi_price.index = bmi['Country']
bmi_price.head(10)
```

```python
bmi_price.plot(kind='barh')
```

```python
from pylab import rcParams

rcParams["figure.figsize"] = 10, 12

bmi_price.plot(kind='barh')

plt.show()
```

```python
bmi_price.sort_values(ascending=True).plot(kind='barh')
plt.show()
```

```python
rcParams['figure.figsize'] = 5, 5
hist_bmi_price = plt.hist(bmi_price, color='blue', ec='black')
print(hist_bmi_price)
plt.show()
```

```python
x = np.linspace(-5, 5, 101)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Function y = x ** 2", 
          fontsize=20, 
          fontweight="bold", 
          color="#22ff00")
plt.show()
```

```python
plt.plot(x, y)
plt.title("y is square of x",
          fontsize=15,
          color="#0000ff")
plt.xlabel("this is x",
           fontsize=10)
plt.ylabel("this is y",
           fontsize=10)
plt.show()
```

```python
y2 = x**3
```

```python
plt.plot(x, y)
plt.plot(x, y2)

plt.title("Graph",
         fontsize=20,
         color="#ff0000")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.axis([-5, 5, -8, 8])
plt.legend(labels=["x^2", "x^3"])

plt.show()
```
